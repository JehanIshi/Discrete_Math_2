**ANSWER:**

To show that the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$, let's calculate $E[Y_i]$, the expected value of $Y_i$.

The expected value of a random variable $Y_i$ is defined as:

$$
E[Y_i] = \sum_{y_i} y_i \cdot P(Y_i = y_i)
$$

Given that $Y_i$ follows a binomial distribution with parameters $n$ and $p_i$, the probability mass function $P(Y_i = y_i)$ is:

$$
P(Y_i = y_i) = \binom{n}{y_i} p_i^{y_i} (1 - p_i)^{n - y_i}
$$

Substituting this into the formula for the expected value:

$$
E[Y_i] = \sum_{y_i} y_i \cdot \binom{n}{y_i} p_i^{y_i} (1 - p_i)^{n - y_i}
$$

This is the expected value of a binomial distribution, which is $n \cdot p_i$. Therefore, $E[Y_i] = n \cdot p_i$.

Now, since $Y_1, Y_2, \ldots, Y_k$ are independent random variables, the expected value of their sum is the sum of their expected values:

$$
E[Y] = E[Y_1 + Y_2 + \ldots + Y_k] = E[Y_1] + E[Y_2] + \ldots + E[Y_k]
$$

Substituting $E[Y_i] = n \cdot p_i$ for each $Y_i$:

$$
E[Y] = n \cdot p_1 + n \cdot p_2 + \ldots + n \cdot p_k = n \cdot (p_1 + p_2 + \ldots + p_k) = n \cdot 1 = n
$$

Thus, the expected value $E[Y] = n \cdot p_i$.

### Question 2:

Given the data:

\[
\begin{align*}
\text{Age} & \quad \text{Proportion} \\
18-24 & \quad 0.18 \\
25-34 & \quad 0.23 \\
35-44 & \quad 0.16 \\
45-64 & \quad 0.27 \\
65-100 & \quad 0.16 \\
\end{align*}
\]

If 500 adults are sampled randomly, we want to find the probability that the sample contains 100 persons between 18 and 24, 200 between 25 and 34, and 200 between 45 and 64.

We use the multinomial probability formula:

$$
P(y_1, y_2, y_3, y_4, y_5) = \frac{n!}{y_1!y_2!y_3!y_4!y_5!} p_1^{y_1} p_2^{y_2} p_3^{y_3} p_4^{y_4} p_5^{y_5}
$$

where $n$ is the total sample size (500 in this case), $y_i$ is the number of individuals in the $i$th age group, and $p_i$ is the proportion of individuals in the $i$th age group.

For the expected value to obtain a person aged 65 and above, we simply multiply the total sample size by the proportion of individuals in that age group:

$$
\text{Expected value for 65 and above} = 500 \times 0.16
$$
